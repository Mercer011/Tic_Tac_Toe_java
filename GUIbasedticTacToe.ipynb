```python
import pygame
import sys

# Initialize Pygame
pygame.init()

# Constants
WIDTH, HEIGHT = 400, 400
LINE_WIDTH = 8
BOX_PADDING = 15  # Added padding
WHITE = (255, 255, 255)
LINE_COLOR = (89, 55, 111)
CIRCLE_COLOR = (200, 0, 0)
CROSS_COLOR = (0, 0, 290)
ORANGE = (255, 165, 0)  # Orange color

# Create the game window
window = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Tic-Tac-Toe")

# Game variables
board = [["" for _ in range(3)] for _ in range(3)]
player_turn = "X"
game_over = False
winner = None

# Game Loop
while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()

        if not game_over:
            if event.type == pygame.MOUSEBUTTONDOWN:
                col = event.pos[0] // (WIDTH // 3)
                row = event.pos[1] // (HEIGHT // 3)
                if board[row][col] == "":
                    board[row][col] = player_turn
                    if player_turn == "X":
                        player_turn = "O"
                    else:
                        player_turn = "X"

        # Check for a win
        for i in range(3):
            if board[i][0] == board[i][1] == board[i][2] != "":
                winner = board[i][0]
                game_over = True
            if board[0][i] == board[1][i] == board[2][i] != "":
                winner = board[0][i]
                game_over = True
        if board[0][0] == board[1][1] == board[2][2] != "":
            winner = board[0][0]
            game_over = True
        if board[0][2] == board[1][1] == board[2][0] != "":
            winner = board[0][2]
            game_over = True

        # Check for a draw
        if all(row.count("") == 0 for row in board) and winner is None:
            game_over = True

    # Draw the game board
    window.fill(WHITE)
    for x in range(1, 3):
        pygame.draw.line(window, LINE_COLOR, (0, x * HEIGHT // 3), (WIDTH, x * HEIGHT // 3), LINE_WIDTH)
        pygame.draw.line(window, LINE_COLOR, (x * WIDTH // 3, 0), (x * WIDTH // 3, HEIGHT), LINE_WIDTH)

    # Draw X's and O's
    for row in range(3):
        for col in range(3):
            if board[row][col] == "X":
                line_width = 10  # Increase the line width for "X"
                x_left = col * WIDTH // 3 + BOX_PADDING + line_width // 2  # Added padding
                x_right = (col + 1) * WIDTH // 3 - BOX_PADDING - line_width // 2  # Added padding
                x_top = row * HEIGHT // 3 + BOX_PADDING + line_width // 2  # Added padding
                x_bottom = (row + 1) * HEIGHT // 3 - BOX_PADDING - line_width // 2  # Added padding
                pygame.draw.line(window, CROSS_COLOR, (x_left, x_top), (x_right, x_bottom), line_width)
                pygame.draw.line(window, CROSS_COLOR, (x_right, x_top), (x_left, x_bottom), line_width)
            elif board[row][col] == "O":
                box_padding = 10  # Added padding
                pygame.draw.circle(window, CIRCLE_COLOR, (col * WIDTH // 3 + WIDTH // 6, row * HEIGHT // 3 + HEIGHT // 6),
                                   min(WIDTH // 6 - BOX_PADDING, HEIGHT // 6 - BOX_PADDING), LINE_WIDTH)

    # Display game over message
    if game_over:
        font = pygame.font.Font(None, 36)
        if winner:
            text = f"Player {winner} wins!"
        else:
            text = "It's a draw!"
        
        text_color = ORANGE  # Orange color
        
        text_render = font.render(text, True, text_color)
        window.blit(text_render, (WIDTH // 2 - text_render.get_width() // 2, HEIGHT // 2 - text_render.get_height() // 2))

    pygame.display.update()
```
